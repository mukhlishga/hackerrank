```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'slowestKey' function below.
#
# The function is expected to return a CHARACTER.
# The function accepts 2D_INTEGER_ARRAY keyTimes as parameter.
#

abcList = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

def slowestKey(keyTimes):
    # Write your code here
    slowestKey = keyTimes[0][0]
    slowestTime = keyTimes[0][1]
    
    for i,_ in enumerate(keyTimes):
        if i == 0:
            start = 0
        else:
            start = keyTimes[i-1][1]
        end = keyTimes[i][1]
        if (end - start) > slowestTime:
            slowestTime = end - start
            slowestKey = keyTimes[i][0]
    
    return abcList[slowestKey]
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    keyTimes_rows = int(input().strip())
    keyTimes_columns = int(input().strip())

    keyTimes = []

    for _ in range(keyTimes_rows):
        keyTimes.append(list(map(int, input().rstrip().split())))

    result = slowestKey(keyTimes)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
2. count between (pass 9/14 too slow)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'countBetween' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY arr
#  2. INTEGER_ARRAY low
#  3. INTEGER_ARRAY high
#

def countBetween(arr, low, high):
    # Write your code here
    result = [0 for i in low]
    
    for i in range(len(low)):
        count = 0
        for num in arr:
            if num >= low[i] and num <= high[i]:
                count += 1
        result[i] = count
    
    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    arr_count = int(input().strip())

    arr = []

    for _ in range(arr_count):
        arr_item = int(input().strip())
        arr.append(arr_item)

    low_count = int(input().strip())

    low = []

    for _ in range(low_count):
        low_item = int(input().strip())
        low.append(low_item)

    high_count = int(input().strip())

    high = []

    for _ in range(high_count):
        high_item = int(input().strip())
        high.append(high_item)

    result = countBetween(arr, low, high)

    fptr.write('\n'.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
3. Collision course (pass all)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'collision' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER_ARRAY speed
#  2. INTEGER pos
#

def collision(speed, pos):
    # Write your code here
    count = 0
    
    for i in range(0, pos):
        if speed[i] > speed[pos]:
            count += 1
    
    for i in range(pos, len(speed)):
        if speed[i] < speed[pos]:
            count += 1
        
    return count

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    speed_count = int(input().strip())

    speed = []

    for _ in range(speed_count):
        speed_item = int(input().strip())
        speed.append(speed_item)

    pos = int(input().strip())

    result = collision(speed, pos)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
4. Perfect substring (pass 4/8 too slow)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'perfectSubstring' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. STRING s
#  2. INTEGER k
#

def check(arr, k):
    myDict = {}
    for i in arr:
        if i not in myDict.keys():
            myDict[i] = 1
        else:
            myDict[i] += 1
    
    status = True
    for values in myDict.values():
        if values != k:
            status = False
            
    return status

def perfectSubstring(s, k):
    # Write your code here
    count = 0
    for i in range(len(s)+1):
        for j in range(len(s)+1):
            if j > i and check(s[i:j],k):
                count += 1
    return count
        
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    k = int(input().strip())

    result = perfectSubstring(s, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys


#
# Complete the 'countSentences' function below.
#
# The function is expected to return a LONG_INTEGER_ARRAY.
# The function accepts following parameters:
#  1. STRING_ARRAY wordSet
#  2. STRING_ARRAY sentences
#

def isAnagram(word1, word2):
    return sorted(word1) == sorted(word2)

def listToString(s):
    return ' '.join(map(str, s))
    
def countSentences(wordSet, sentences):
    # Write your code here
    from itertools import permutations
    result = [0 for i in sentences]
    
    for i,sentence in enumerate(sentences):
        count = 0
        sentenceList = sentence.split()
        for j,word in enumerate(sentenceList):
            for wordOption in wordSet:
                if isAnagram(word, wordOption) and word != wordOption:
                    count += 1
        
        temp = [k for k in range(count)]
        perm = permutations(temp, count)
        output = 0
        for k in list(perm):
            output += 1 
                       
        result[i] = output*2
    
    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    wordSet_count = int(input().strip())

    wordSet = []

    for _ in range(wordSet_count):
        wordSet_item = input()
        wordSet.append(wordSet_item)

    sentences_count = int(input().strip())

    sentences = []

    for _ in range(sentences_count):
        sentences_item = input()
        sentences.append(sentences_item)

    result = countSentences(wordSet, sentences)

    fptr.write('\n'.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```
