### Approach 1: Functional Programming

```python
def compareShipMissile(SLoc, MLoc):
    assert type(SLoc) == str and type(MLoc) == str
    
    SLoc = SLoc.split(":")
    MLoc = MLoc.split(":")
    
    dead = []
    alive = []
    miss = []
    output = []
    
    for S in SLoc:
        if S in MLoc:
            dead.append(S)
        else:
            alive.append(S)
            
    for M in MLoc:
        if M not in SLoc:
            miss.append(M)
    
    output.append(dead)
    output.append(alive)
    output.append(miss)
    
    return output
```

```python
def printResult(M, dead, alive, miss):
    assert type(dead) == list and type(alive) == list and type(miss) == list
    
    arr = [['_' for i in range(M)] for j in range(M)]
    
    for i in dead:
        arr[int(i[0])][int(i[2])] = 'X'
    for i in alive:
        arr[int(i[0])][int(i[2])] = 'B'
    for i in miss:
        arr[int(i[0])][int(i[2])] = 'O'
    
    for i in range(len(arr)):
        for j in range(len(arr)):
            print(arr[i][j], end=' ')
        print()
```

```python
def finalResult(dead1, dead2):
    assert type(dead1) == list and type(dead2) == list
    
    P1hit = len(dead2)
    P2hit = len(dead1)
    
    print('P1:' + str(P1hit))
    print('P2:' + str(P2hit))
    
    if P1hit > P2hit:
        print('P1 win')
    elif P1hit < P2hit:
        print('P2 win')
    else:
        print('It is a draw')
```

```python
def main(M, S, P1S, P2S, T, P1M, P2M):
    assert M > 0 and M < 10, "M must be 0 < M < 10"
    assert S > 0 and S <= (M**2)/2, "S must be 0 < S <= M^2/2"
    assert T > 0 and T < 100, "M must be 0 < T < 100"
    
    dead1, alive1, miss1 = compareShipMissile(P1S, P2M)
    dead2, alive2, miss2 = compareShipMissile(P2S, P1M)
    print('player1')
    printResult(M, dead1, alive1, miss1)
    print()
    print('player2')
    printResult(M, dead2, alive2, miss2)
    print()
    finalResult(dead1, dead2)
```

```python
# Case 1 - draw
M = 5
S = 5
P1S = '1,1:2,0:2,3:3,4:4,3'
P2S = '0,1:2,3:3,0:3,4:4,1'
T = 5
P1M = '0,1:4,3:2,3:3,1:4,1'
P2M = '0,1:0,0:1,1:2,3:4,3'

main(M, S, P1S, P2S, T, P1M, P2M)
```

```python
# Case 2 - player 2 win
M = 5
S = 5
P1S = '1,1:2,0:2,3:3,4:4,3'
P2S = '0,1:2,3:3,0:3,4:4,1'
T = 5
P1M = '0,0:4,3:2,3:3,1:4,1'
P2M = '0,1:0,0:1,1:2,3:4,3'

main(M, S, P1S, P2S, T, P1M, P2M)
```

### Approach 2: Object Oriented Programming

```python
class Player():
    def __init__(self, M, S, SLoc, T, MLoc):
        assert M > 0 and M < 10, "M must be 0 < M < 10"
        assert S > 0 and S <= (M**2)/2, "S must be 0 < S <= M^2/2"
        assert T > 0 and T < 100, "M must be 0 < T < 100"
        assert type(SLoc) == str and type(MLoc) == str
        self.M = M
        self.S = S
        self.SLoc = SLoc.split(":")
        self.T = T
        self.MLoc = MLoc.split(":")
    
    def ShipLoc(self):
        arr = [['_' for i in range(self.M)] for j in range(self.M)]
        for i in self.SLoc:
            arr[int(i[0])][int(i[2])] = 'S'
        return arr
    
    def MisLoc(self):
        arr = [['_' for i in range(self.M)] for j in range(self.M)]
        for i in self.MLoc:
            arr[int(i[0])][int(i[2])] = 'T'
        return arr
```

```python
class Game():
    def __init__(self, player1, player2):
        assert isinstance(player1, Player) and isinstance(player1, Player)
        self.player1 = player1
        self.player2 = player2
        self.player1.dead = []
        self.player1.alive = []
        self.player1.miss = []
        self.player2.dead = []
        self.player2.alive = []
        self.player2.miss = []
    
    def compareShipMissile(self):
        #stats for player 1
        for S in self.player1.SLoc:
            if S in self.player2.MLoc:
                self.player1.dead.append(S)
            else:
                self.player1.alive.append(S)
                
        for M in self.player2.MLoc:
            if M not in self.player1.SLoc:
                self.player1.miss.append(M)
                
        #stats for player 2
        for S in self.player2.SLoc:
            if S in self.player1.MLoc:
                self.player2.dead.append(S)
            else:
                self.player2.alive.append(S)

        for M in self.player1.MLoc:
            if M not in self.player2.SLoc:
                self.player2.miss.append(M)
                
    def printResult(self):
        #printing player 1 grid
        grid1 = [['_' for i in range(self.player1.M)] for j in range(self.player1.M)]

        for i in self.player1.dead:
            grid1[int(i[0])][int(i[2])] = 'X'
        for i in self.player1.alive:
            grid1[int(i[0])][int(i[2])] = 'B'
        for i in self.player1.miss:
            grid1[int(i[0])][int(i[2])] = 'O'

        print('player1')
        for i in range(len(grid1)):
            for j in range(len(grid1)):
                print(grid1[i][j], end=' ')
            print()
            
        print()
        #printing player 2 grid
        
        grid2 = [['_' for i in range(self.player2.M)] for j in range(self.player2.M)]

        for i in self.player2.dead:
            grid2[int(i[0])][int(i[2])] = 'X'
        for i in self.player2.alive:
            grid2[int(i[0])][int(i[2])] = 'B'
        for i in self.player2.miss:
            grid2[int(i[0])][int(i[2])] = 'O'

        print('player2')
        for i in range(len(grid2)):
            for j in range(len(grid2)):
                print(grid2[i][j], end=' ')
            print()
            
        print()
        #printing the final score
        
        P1hit = len(self.player2.dead)
        P2hit = len(self.player1.dead)

        print('P1:' + str(P1hit))
        print('P2:' + str(P2hit))

        if P1hit > P2hit:
            print('P1 win')
        elif P1hit < P2hit:
            print('P2 win')
        else:
            print('It is a draw')
```

```python
# Case 1 - draw
M = 5
S = 5
P1S = '1,1:2,0:2,3:3,4:4,3'
P2S = '0,1:2,3:3,0:3,4:4,1'
T = 5
P1M = '0,1:4,3:2,3:3,1:4,1'
P2M = '0,1:0,0:1,1:2,3:4,3'

player1 = Player(M, S, P1S, T, P1M)
player2 = Player(M, S, P2S, T, P2M)

game1 = Game(player1, player2)
game1.compareShipMissile()
game1.printResult()
```

```python
# Case 2 - player 2 win
M = 5
S = 5
P1S = '1,1:2,0:2,3:3,4:4,3'
P2S = '0,1:2,3:3,0:3,4:4,1'
T = 5
P1M = '0,0:4,3:2,3:3,1:4,1'
P2M = '0,1:0,0:1,1:2,3:4,3'

player1 = Player(M, S, P1S, T, P1M)
player2 = Player(M, S, P2S, T, P2M)

game2 = Game(player1, player2)
game2.compareShipMissile()
game2.printResult()
```
